# Imagine you have to present a report showing the top 10 most fuel efficient vehicle manufacturers for 2016. 

# We know that we have the data we need in a CSV file containing data from several years.


# What do do?

# 1. Open and read the data set.

```python
import pandas as pd
from IPython.display import HTML
```

```python
data = pd.read_csv('vehicles.csv')
data
```

# 2. We will need to filter the data just for the year 2016

```python
year = 2016
filtered = data[data['Year']==year]
filtered
```

# In our data set, there is a Make field that tells us the vehicle's manufacturer and a Combined MPG field that tells us the fuel efficiency for every vehicle. 

# To get the results we will need to:

# 3. Group by manufacturer and Average Combined MPG.

```python
grouped = filtered.groupby('Make').agg({'Combined MPG':'mean'}).reset_index()
grouped
```

# 4. Sort by average Combined MPG (in descending order).

```python
grouped = filtered.groupby('Make').agg({'Combined MPG':'mean'}).reset_index()
results = grouped.sort_values('Combined MPG', ascending=False)
results
```

# 5. Keep just the top 10 results.

```python
results_10 = grouped.sort_values('Combined MPG', ascending=False).head(10)
results_10
```

# 6.Create a bar chart to show the differences between manufacturer fuel efficiency.

```python
import matplotlib.pyplot as plt
import seaborn as sns
 
title = 'Top 10 Manufacturers by Fuel Efficiency ' + str(year)
 
fig, ax = plt.subplots(figsize=(15,8))
barchart = sns.barplot(data=results_10, x='Make', y='Combined MPG')
plt.title(title + "\n", fontsize=16)
```

# 7. Export the chart as a image file to a shared directory where you and your team can see it.

```python
fig = barchart.get_figure()
fig.savefig(title + '.png')
```

# 8. Automating the Pipeline
### Now that we have written the code to produce this analysis once, we will want to automate it so that we can run it again next time without having to modify the code. 

### i) To do this, we are going to organize the code by wrapping it into functions according to the step in the pipeline. 

### ii) We will also make the "year" and "title" variables global (because they change from year to year)

### iii) The modified code should like this.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
 
year = int(input('Enter the year: '))
title = 'Top 10 Manufacturers by Fuel Efficiency ' + str(year)
 
def acquire():
    data = pd.read_csv('vehicles.csv')
    return data
 
def wrangle(df):
    filtered = data[data['Year']==year]
    return filtered
 
def analyze(df):
    grouped = filtered.groupby('Make').agg({'Combined MPG':'mean'}).reset_index()
    results = grouped.sort_values('Combined MPG', ascending=False).head(10)
    return results
 
def visualize(df):
    fig, ax = plt.subplots(figsize=(15,8))
    barchart = sns.barplot(data=results, x='Make', y='Combined MPG')
    plt.title(title + "\n", fontsize=16)
    return barchart
 
def save_viz(barchart):
    fig = barchart.get_figure()
    fig.savefig(title + '.jpeg')
 
if __name__ == '__main__':
    data = acquire()
    filtered = wrangle(data)
    results = analyze(filtered)
    barchart = visualize(results)
    save_viz(barchart)
```
